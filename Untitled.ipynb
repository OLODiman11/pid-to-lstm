```python
def generate_temperature_setpoints(initial_rest, initial_temp, min_interval, max_interval, min_temp, max_temp, count):
    start = 0
    end = initial_rest
    Tsp = np.zeros(count) * initial_temp
    while end <= count:
        start = end
        end += random.randint(min_interval, max_interval) 
        Tsp[start:end] = random.randint(min_temp, max_temp)
    return Tsp
        
def set_temperature_setpoints(x, y, count):
    Tsp = np.zeros(count)
    for i in range(len(x) - 1):
        Tsp[x[i]:x[i + 1]] = y[i]
    Tsp[x[len(x) - 1]:count] = y[len(y) - 1]
    return Tsp

def plot(data, x_label, y_label, label_size=14, ticks_size=12):
    plt.plot(data)
    plt.xlabel(x_label, size=label_size)
    plt.ylabel(y_label, size=label_size)
    plt.xticks(size=ticks_size)
    plt.yticks(size=ticks_size)
```

```python
Tsp = set_temperature_setpoints(
    [0, 31, 446, 1036, 1294, 1553, 1889, 2413, 2972, 3382, 3723, 4297, 4592, 5020, 5303],
    [23.67, 41, 57, 35, 67, 52, 46, 58, 57, 69, 41, 54, 65, 52, 65], 60.0*run_time)
plot(Tsp, 'Time', r'Temp SP ($^oC$)')
```

```python
def run_tclab(run_time, Tsp, active_cont, passive_cont = None, emulator=False, speedup=60):
    
    loops = int(60.0*run_time)

    T1 = np.zeros(loops)
    Q1 = np.zeros(loops)
    tm = np.zeros(loops)
    
    if emulator:
        TCLab = tclab.setup(connected=False, speedup=speedup)
    with TCLab() as lab:

        start_time = 0
        prev_time = -1

        i = 0
        for clock in tclab.clock(run_time * 60):
            if i >= 60.0 * run_time:
                break

            t = clock
            dt = t - prev_time
            prev_time = t
            tm[i] = t - start_time

            T1[i] = lab.T1
            passive_Q = 0 
            if passive_cont is not None:
                passive_Q = passive_cont.get_output(Tsp[i], T1[i], dt)
            Q1[i] = active_cont.get_output(Tsp[i], T1[i], dt)
            lab.Q1(Q1[i])
            print(f'{tm[i]:6.1f} {T1[i]:6.2f} {Q1[i]:6.2f} {passive_Q:6.2f}')
            
            i += 1
            
        return tm, T1, Q1
    
```

```python
tm, T1, Q1 = run_tclab(Tsp, emulator=True, speedup=60)
```

```python
# Save csv file
data = pd.DataFrame()
data['Q1'] = Q1
data['T1'] = T1
data['Tsp'] = Tsp

# Plot
plt.plot(data['Q1'],'b-',label='$Q_1$ (%)')
plt.plot(data['T1'],'r-',label='$T_1$ $(^oC)$')
plt.plot(data['Tsp'],'k-',label='SP $(^oC)$')
plt.legend()
```

```python
import import_ipynb
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pickle

# For scaling, feature selection
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_selection import SelectKBest, f_regression
from sklearn.model_selection import train_test_split 

# For LSTM model

from keras.models import Sequential
from keras.layers import LSTM, Dropout, Dense
from keras.callbacks import EarlyStopping
from tqdm.keras import TqdmCallback
from keras.models import load_model
from lstmcontroller import LstmController
```

```python
def select_best_features(X, y, threshold = 0.05):
    bestfeatures = SelectKBest(score_func=f_regression, k='all')
    fit = bestfeatures.fit(X,y)
    scores = fit.scores_
    total_score = sum(scores)
    
    features = list()
    
    for i in range(len(X.columns)):
        if scores[i] >= threshold * total_score:
            features.append(X.columns[i])
    
    plt.bar(x=X.columns,height=fit.scores_ / total_score);
    
    return X[features]
```

```python
train_data = pd.read_csv('PID_train_data.csv')
train_data['err'] = train_data['Tsp'] - train_data['T1']

y = np.ravel(train_data[['Q1']])
X = select_best_features(train_data[['T1','Tsp','err']], y, 0.1)

```

```python
def create_train_and_test_data(X, y, s_x, s_y, window, test_size=0.2, shuffle=False):

        Xs = s_x.fit_transform(X)
        ys = s_y.fit_transform(y)

        X_lstm = []
        y_lstm = []
        for i in range(window, len(X)):
            X_lstm.append(Xs[i-window:i])
            y_lstm.append(ys[i])

        X_lstm, y_lstm = np.array(X_lstm), np.array(y_lstm)
        return train_test_split(X_lstm, y_lstm, test_size=test_size, shuffle=shuffle)
```

```python
s_x = MinMaxScaler()
s_y = MinMaxScaler()
Xtrain, Xtest, ytrain, ytest = create_train_and_test_data(X.values, y.reshape(-1, 1), s_x, s_y, 15)
lstm = LstmController(s_x, s_y, Xtrain.shape, 2, 15, 0.1, 100)
lstm.train_model(Xtrain, ytrain, 100)
lstm.save_model('pid_emulate.h5', 'model_params.pkl')
```

```python
yp = lstm.predict(Xtest)
print(Xtest)

Xtest_us = lstm.s_x.inverse_transform(Xtest[:,-1,:])
ytest_us = lstm.s_y.inverse_transform(ytest)

sp = Xtest_us[:,0]
pv = Xtest_us[:,0] + Xtest_us[:,1]

plt.plot(sp,'k-',label='$SP$ $(^oC)$')
plt.plot(pv,'r-',label='$T_1$ $(^oC)$')
plt.plot(ytest_us,'b-',label='$Q_{PID}$ (%)')
plt.plot(yp,'g-',label='$Q_{LSTM}$ (%)')
plt.legend(fontsize=12,loc='lower right')
plt.xlabel('Time',size=14)
plt.ylabel('Value',size=14)
plt.xticks(size=12)
plt.yticks(size=12);
```
