```python
import import_ipynb
from pidcontroller import PidController
import functions as fn
import matplotlib.pyplot as plt
import pandas as pd
```

```python
run_time = 90.0 * 60
speedup = 30

Kc   = 6.0
tauI = 75.0 # sec
tauD = 0.0  # sec
out_min = 0
out_max = 100
pid = PidController(Kc, Kc / tauI, -Kc * tauD, out_min, out_max)
```

```python
Tsp = fn.set_temperature_setpoints(
    [0, 31, 446, 1036, 1294, 1553, 1889, 2413, 2972, 3382, 3723, 4297, 4592, 5020, 5303],
    [23.67, 41, 57, 35, 67, 52, 46, 58, 57, 69, 41, 54, 65, 52, 65], int(run_time))
fn.plot(Tsp, 'Time', r'Temp SP ($^oC$)')
plt.savefig('SP_profile.png');
```

```python
tm, T1, Q1, *_ = fn.run_tclab(run_time, Tsp, pid, emulator=True, speedup=speedup)
```

```python
data = pd.DataFrame()
data['Q1'] = Q1
data['T1'] = T1
data['Tsp'] = Tsp
data.to_csv('PID_train_data.csv',index=False)
```

```python
fn.multiplot({
    '$Q_1$ (%)': Q1,
    '$T_1$ $(^oC)$': T1,
    'SP $(^oC)$': Tsp
})
plt.savefig('PID_train.png');
```
